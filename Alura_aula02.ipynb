```python
cod_compostos = dados['composto'].value_counts().index[0:5] #primeiros 5 elementos (index)
cod_compostos
```

```python
#dados.query('composto == cod_compostos') #da erro
dados.query('composto in @cod_compostos') #@ esta declarado anteriomente / in √© igual ==
```

```python
import seaborn as sns
import matplotlib.pyplot as ply

plt.figure(figsize=(8,6)) #Tamanho da imagem
sns.set() #visual seaborn
ax = sns.countplot(x = 'composto',data=dados.query('composto in @cod_compostos'),palette='winter')
ax.set_title('Top 5 Compostos')
plt.show()
```

```python
#DESAFIO 01 e 02

import seaborn as sns
import matplotlib.pyplot as ply

plt.figure(figsize=(8,6)) #Tamanho da imagem
sns.set() #visual seaborn

ax = sns.countplot(x = 'composto',data=dados.query('composto in @cod_compostos'),order=cod_compostos,palette="rocket")

ax.set_title('Top 5 Compostos', fontsize=20)
ax.set_xlabel('Composto', fontsize=15)
ax.set_ylabel('Quantidade', fontsize=15)
plt.show()
```

```python
#DESAFIO 03

fig, ax = plt.subplots(1,3, figsize = ( 20 , 5 ))
dados.loc[:,'c0':'c99'].describe().T['mean'].hist(bins=50,ax=ax[0],color ='#1E73BE', histtype='stepfilled').set_title("M√©dia dos tipos celulares")
dados.loc[:,'c0':'c99'].describe().T['min'].hist(bins=50,ax=ax[1],color ='#C850B0',histtype='bar').set_title("M√≠nimo dos tipos celulares")
dados.loc[:,'c0':'c99'].describe().T['max'].hist(bins=50,ax=ax[2],color ='#660A60',histtype='step').set_title("M√°ximo dos tipos celulares")
fig.suptitle('Histogramas dos tipos celulares', fontsize=20, color="#000C66")

fig.show()
```

###Desafio 04

Describe() 

O m√©todo descreve as informa√ß√µes estat√≠sticas de retorno dos dados do dataframe ou dados em s√©rie (de fato, este m√©todo retorna um dataframe). Esta informa√ß√£o inclui o n√∫mero de amostras, o valor m√©dio, o desvio padr√£o, o valor m√≠nimo, o valor m√°ximo, a mediana e os valores correspondentes aos percentis de 25% e 75%.



```python
#Desafio 06

fig, ax = plt.subplots(3, 3, figsize=(20,10), sharex=True, sharey=True) 


sns.histplot(dados['c0'], ax=ax[0,0],bins=100,color='#FFCA72', element="poly")
sns.histplot(dados['c10'], ax=ax[0,1],bins=100,color='#FF4500', kde=True, element="step")
sns.histplot(dados['c20'], ax=ax[0,2],bins=100,color='#DAD870', kde=True)
sns.histplot(dados['c30'], ax=ax[1,0],bins=100,color='#FF9636', kde=True, fill=False)
sns.histplot(dados['c40'], ax=ax[1,1],bins=100,color='#AE388B',kde=True)
sns.histplot(dados['c50'], ax=ax[1,2],bins=100,color='#FF5C4D', kde=True)
sns.histplot(dados['c60'], ax=ax[2,0],bins=100,color='#B8390E', kde=True)
sns.histplot(dados['c70'], ax=ax[2,1],bins=100,color='#DC4731', kde=True)
sns.histplot(dados['c80'], ax=ax[2,2],bins=100,color='#FB6090')

fig.suptitle('Distribui√ß√£o de 9 Vari√°veis Com Prefixo c', fontsize=20, color='#2AB67B')    

plt.show()
```

```python
#Desafio 06

plt.figure(figsize=(10,8))
ax = sns.boxplot(y='g0', x='tratamento', hue='dose', data=dados, palette="husl")
ax.set_title('Valores de g0 - com droga e com controle', fontsize = 20, color='#2AB67B')
ax.set_xlabel('Tratamento', fontsize = 15)
ax.set_ylabel('Valores', fontsize = 15)
plt.show()
```
